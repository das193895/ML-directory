```python
df_dummies = pd.get_dummies(df, columns=["sales","salary"])
```

```python
df_dummies.head()
```

```python
df_dummies.columns
```

```python
## Dependent and Independent variable
X = df_dummies.drop(["left"],axis = 1)
y = df_dummies[["left"]]
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'X' is your feature matrix and 'y' is your target variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=123)

# Now, X_train and y_train are the training features and labels, and X_test and y_test are the testing features and labels
```

```python
from imblearn.over_sampling import SMOTE
smk = SMOTE()
x_res,y_res=smk.fit_resample(X_train,y_train)
```

```python
## x_res and y_res are new training data
```

```python
from sklearn.model_selection import cross_val_predict, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report

# X_res: Features, y_res: Target variable

# Function to perform 5-Fold Cross-Validation and plot classification report
def cross_validate_and_plot(model, x_res, y_res, title):
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    y_pred = cross_val_predict(model, x_res, y_res, cv=cv)

    # Plot classification report
    report = classification_report(y_res, y_pred, target_names=['Class 0', 'Class 1'])
    print(f"\nClassification Report for {title}:\n{report}")

    # Plot confusion matrix if needed
    # ...

# 5.1 Logistic Regression
logistic_model = LogisticRegression()
cross_validate_and_plot(logistic_model, x_res, y_res, 'Logistic Regression')

# 5.2 Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
cross_validate_and_plot(rf_model, x_res, y_res, 'Random Forest Classifier')

# 5.3 Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
cross_validate_and_plot(gb_model, x_res, y_res, 'Gradient Boosting Classifier')

# Additional: You can customize the models and parameters based on your specific requirements.
```

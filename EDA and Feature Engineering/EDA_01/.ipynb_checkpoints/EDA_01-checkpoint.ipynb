```python
df_country = pd.read_excel("Country-code.xlsx")
df_country.head()
```

```python
df.columns
```

```python
## we can see that country code is an existing column in the df data frame so we can merge the country code data frame to the df data frame
```

```python
df_new = pd.merge(df,df_country,on="Country Code",how="left")
```

```python
df_new.head()
```

```python
df_new.dtypes
```

```python
df_new["Country"].value_counts()
```

```python
country_array = df_new["Country"].value_counts().index
country_array
```

```python
country_values = df_new["Country"].value_counts().values
country_values
```

```python
# pie chart for top 3 countries using zomato
plt.pie(country_values[:3],labels = country_array[:3],autopct="%1.2f%%")
```

```python
df_new.columns
```

```python
ratings = df_new.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:"rating count"})
ratings
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(y = ratings["rating count"] , x = ratings["Aggregate rating"] )
```

```python
sns.countplot(x = ratings["Rating color"])
```

```python
# how many Zero ratings given by which country
df_new[df_new["Aggregate rating"] == 0].groupby("Country").size()
```

```python
df_new.columns
```

```python
# Which currency is used by which country
df_new[["Country","Currency"]].groupby(["Country","Currency"]).size().reset_index()  
```

```python
# countries having online delivery
df_new[df_new["Has Online delivery"] == 'Yes'].groupby("Country").size()
```

```python
df_new[["Has Online delivery","Country"]].groupby(["Has Online delivery","Country"]).size().reset_index()
```

```python
## Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import math
```

```python
## function for calculating euclidian distance between 2 n-dimensional inputs

def calc_euclidian_distance(input_1,input_2):  ## here input_1 and input_2 are vectors of same dimensions
    
    squared_distance = 0
    for (x,y) in zip(input_1,input_2):
        squared_distance = squared_distance + (x-y)**2  ## looping through all the the data points in the 2 input vectors and calculating the squared_distance
        
    final_distance = math.sqrt(squared_distance)  ## taking under_root
    
    return final_distance   ## return the calculated euclidian distance
        
```

```python
## function for sampling 100 d-dimensional points from the unit cube

## give the dimension of input variable you want to sample

## returns a column vector withh 100 points of specified dimension 

def sampling(dimension):
    # define a column vector to store all the data points
    storage = np.zeros((100,1))
    
    storage = np.random.rand(100,dimension)  ## sampling random points of the given dimension
    
    return storage
```

```python
def calc_stats():
    ## define a matrix to store distances and standard deviations of each points from each other
    distance_storage = np.zeros((100,100))
    
    dimension_storage = [2**i for i in range(11)]
    average_storage = np.zeros(11)
    std_storage = np.zeros(11)
    
    for d in range(0,11):
        
        store_points = sampling(2**d);
        
        for i in range(len(store_points)):
            for j in range(len(store_points)):
                distance_storage[i][j] = calc_euclidian_distance(store_points[i,:],store_points[j,:])**2            
                       
        average_storage[d] = np.mean(distance_storage)
        std_storage[d] = np.std(distance_storage)
        
        
    ## plotting
    
    plt.figure(figsize = (5,5))
    plt.plot(dimension_storage, average_storage , marker = "o")
    plt.xlabel('2^dimensions')
    plt.ylabel('averages')
    plt.title('plotting between  average squared Euclidean distance between all pairs of points as a function of d')
    
    plt.figure(figsize = (5,5))
    plt.plot(dimension_storage,std_storage,marker = "o")
    plt.xlabel('2^dimensions')
    plt.ylabel('standard_deviations')
    plt.title('plotting between standard deviation of the squared Euclidean distances as a function of d')
    
    
    
    return dimension_storage,average_storage,std_storage
```

```python
calc_stats()
```

```python
## Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import math
import scipy.io
```

## Data Gathering and preprocessing

```python
# Load MATLAB file
mat_data = scipy.io.loadmat('house_data.mat')

train_X_arr = mat_data["Xtrain"]
test_X_arr = mat_data["Xtest"]
train_y_arr = mat_data["Ytrain"]
test_y_arr = mat_data["Ytest"]
```

## Root mean square loss function

```python
def computing_cost_fn(feature_matrix, target_var, weight_matrix):
    
    ones_column = np.ones((feature_matrix.shape[0], 1)) 
    X = np.hstack((ones_column, feature_matrix))  # Stacking a column of ones 
    y = target_var

    no_of_samples = X.shape[0]
    
    summation = 0.0;
    
    
    for x,y in zip(X, y):
        y_hat = np.dot(x,weight_matrix)
        summation += (y_hat - y) ** 2
        
    final_cost = summation / (no_of_samples * 2.0)  # calculate final cost

    return final_cost[0]**0.5
```

## Function for calculating the optimum weights using direct solution

```python
def direct_solution(feature_matrix,target_variable):
    
    ones_column = np.ones((feature_matrix.shape[0], 1))
    
    X = np.hstack((ones_column, feature_matrix))
    
    a = np.dot(X.T,X)
    
    a_inv = np.linalg.inv(a)
    
    b = np.dot(a_inv,X.T)
    
    c = np.dot(b,target_variable)
    
    return c
```

## Implement linear Regression

```python
def linear_Regression(testing_features,testing_target_var,feature_matrix, target_var):
    
    ## compute the optimum weights using the direct solution
    
    updated_weight_matrix = direct_solution(feature_matrix,target_var) # calculate optimum weights 
    
    # adding a column of ones to the testing feature matrix 
    ones_column = np.ones((testing_features.shape[0], 1))
    
    X_test = np.hstack((ones_column, testing_features))
    
    no_of_samples = X_test.shape[0]
    
    y_test = testing_target_var
    
    
    # predicting the new y values with updated weight matrix
    y_pred = np.dot(X_test,updated_weight_matrix)
        
    return y_pred,updated_weight_matrix  ## return the predicted y_value vector 
```

```python
predictions,weights = linear_Regression(test_X_arr,test_y_arr,train_X_arr,train_y_arr)
```

```python
computing_cost_fn(train_X_arr, train_y_arr, weights)
```

## Making the plots

```python
# Exclude the intercept term
theta_without_intercept = weights[1:]

# Plot the regression parameters against their indices
plt.plot(np.arange(len(theta_without_intercept)), theta_without_intercept, marker='o')
plt.xlabel('Parameter Index')
plt.ylabel('Parameter Value')
plt.title('Regression Parameters')
plt.grid(True)
plt.show()
```

```python
residuals = (predictions-test_y_arr)  ## calculate residuals
```

```python
# Plot histogram of residuals
plt.hist(residuals, bins=10, edgecolor='black')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.title('Histogram of Residuals')
plt.grid(True)
plt.show()
```
